##Product Rule

If $f(x)$ and $g(x)$ are differentiable in $x$, then their product is differentiable in $x$.


Let $h(x)=f(x)g(x)$


\begin{equation}
h'(x)=f(x)g'(x)+f'(x)g(x)
\end{equation}


**Examples:**
Calculate the derivative of $f(x)g(x)$ and graph in python

\begin{align}
f(x)=x^2 \qquad
g(x)=x^3
\end{align}

```python
syp.var('x')
fx=x**2
gx=x**3
dfxgx=syp.diff(fx*gx, x)
print(dfxgx)
dfxgx.doit()

```

```python
dfx=syp.diff(fx, x)
dgx=syp.diff(gx, x)
dfxgx=fx*dgx+dfx*gx
print(dfxgx)
dfxgx.doit()
```

##Derivatives of Sums, Differences and Constant Multiples

If $f$ and $g$ are differentiable at $x$, and $c$ is any constant, then $f+g$, $f-g$ and $cf$ are differentiable at $x$.


\begin{align}
(f+g)'(x) = f'(x)+g'(x)\\
(f-g)'(x) = f'(x)-g'(x)\\
(cf)'(x) = cf'(x)\\
\end{align}

##Reciprocal Rule 
If $f$ is differentiable in $x$ and $f(x)\neq 0$, then $f(\frac{1}{f})$ is differentiable in $x$.

\begin{align}
\left(\frac{1}{f}\right)'(x)=-\frac{f'(x)}{(f(x))^2}
\end{align}

##Quotient Rule
If $f$ and $g$ are differentiable at $x$ and $g(x)\neq 0$, then their quotient is differentiable at $x$,

\begin{align} 
\frac{d}{dx}\left[\frac{f(x)}{g(x)}\right]=\frac{g(x)f'(x)-f(x)g'(x)}{(g(x))^2}
\end{align} 

**Example:**

Let $f(x)=x$ and $g(x)=x^2+1$ calculate the derivate and graph it in Python.



```python
syp.var('x')
fx=x
gx=x**2+1

dfx_gx=syp.diff(fx/gx, x)
print(dfx_gx)
dfx_gx
```

```python
d1fx=syp.diff(fx,x)
d1gx=syp.diff(gx,x)
dfx_gx=(gx*d1fx-fx*d1gx)/(gx**2)
print(dfx_gx)
dfx_gx
syp.simplify(dfx_gx)
```

##Higher Order Derivatives

 Let $f$ be a differentiable function, then $f'$ is the first derivative of $f$. If $f'(x)$ is also derivable, then $f''(x)$ is called the second derivative of the primitive function $f$. In this way it is possible to obtain derivatives of higher order.

\begin{align}
f'(x) = y' = \frac{dy}{dx} = \frac{d}{dx}f(x)
\end{align}

**Example:**

Calculate the derivatives of order 1,2,3,4 and 5 for the function $f(x)=x^4$.

***Solution:***

\begin{align}
f' (x) = 4x^3 \\
f'' (x) = 12x^2 \\
f''' (x) = 24x\\
f^{(4)} (x) = 24 \\
f^{(5)} (x) = 0 \\
\end{align}



```python
x= syp.symbols('x') 
f = 4*x**3
syp.Derivative(f,x, 5).doit()
```

##Higher Order Partial Derivatives

In the same way as with functions of one variable, it is possible to find the second, third, etc, partial derivatives.

**Order of Partial Derivatives of a function of two variables**. 

Let $z=f(x,y)$

Second derivative with respect to $x$:
\begin{align}
\frac{\partial}{\partial x} \left( \frac{\partial f}{\partial x} \right) =
\frac{\partial^2 f}{\partial x^2}=f_{xx}
\end{align}

Second derivative with respect to $y$:
\begin{align}
\frac{\partial}{\partial y} \left( \frac{\partial f}{\partial y} \right) =
\frac{\partial^2 f}{\partial y^2}=f_{yy}
\end{align}


Derive first with respect to $x$ and then with respect to $y$:
\begin{align}
\frac{\partial}{\partial y} \left( \frac{\partial f}{\partial x} \right) =
\frac{\partial^2 f}{\partial y \partial x}=f_{xy}
\end{align}
Derive first with respect to $y$ and then with respect to $x$:
\begin{align}
\frac{\partial}{\partial x} \left( \frac{\partial f}{\partial y} \right) =
\frac{\partial^2 f}{\partial x \partial y}=f_{yx}
\end{align}

#The Total Derivative

A total derivative of a multivariable function of several variables, each of which is a function of another argument, is the derivative of the function with respect to said argument. it is equal to the sum of the partial derivatives with respect to each variable times the derivative of that variable with respect to the independent variable. For example, given a function $f(x,y,z)$ , and $x,y,z$ being functions of $t$ ,

\begin{align}
\frac{df(x,y,z)}{dt}=\frac {\partial f}{\partial x} \cdot \frac {dx}{dt}+\frac {\partial f}{\partial y} \cdot \frac {dy}{dt}+\frac {\partial f}{\partial z}\cdot \frac {dz}{dt}
\end{align}
The formula for a total derivative is a direct result of the chain rule.

```python
from sympy.abc import x, y, z, t
fxyz= syp.sin(x)*syp.E**(y*z**2)
fxyz
```

```python
dfxyz_x=syp.diff(fxyz,(x,1))
dfxyz_y=syp.diff(fxyz,(y,1))
dfxyz_z=syp.diff(fxyz,(z,1))
```

```python
fxt=t-1; fyt=t**2; fzt=1/t
ft = fxyz.subs(x, fxt) 
ft = ft.subs(y, fyt) 
ft = ft.subs(z, fzt) 
ft.doit()
```

```python
dfdt=syp.diff(ft,(t,1))
dfdt.doit()
```

```python
dfx_t=syp.diff(fxt,(t,1)) 
dfy_t=syp.diff(fyt,(t,1))
dfz_t=syp.diff(fzt,(t,1))
dfxyz_t=dfxyz_x*dfx_t + dfxyz_y*dfy_t + dfxyz_z*dfz_t
dfxyz_t
```

```python
dfxyz_t = dfxyz_t.subs(x, fxt)
dfxyz_t = dfxyz_t.subs(y, fyt)
dfxyz_t = dfxyz_t.subs(z, fzt) 
dfxyz_t
```
