function MateKeys(text, keys, stat) {
    var newgen = [];
    while(newgen.length < populationSize) {
        var key = keys[Math.floor(Math.random()*keys.length)];
        var mate = keys[Math.floor(Math.random()*keys.length)];
        var target = copyKey(key.fitness < mate.fitness ? key : mate);
        var source = key.fitness < mate.fitness ? mate : key;
        var mated = false;
        alphabet.forEach(function(ch) {
            if (source[ch] != target[ch]) {
                var oldvalue = target[ch];
                swapValues(target, source[ch], oldvalue);
                var fitness =  calculateKeyFitness(text, target, stat);
                if (fitness < target.fitness) {
                    target.fitness = fitness;
                    mated = true;
                } else {
                    //second swap swaps back
                    swapValues(target, source[ch], oldvalue);
                }
            }
        });
        newgen.push(target);
    }
    return newgen;
}

function MutateKeys(text, keys, stat) {
    keys.forEach(function(key) {

                key[ch] = key[swap];
                key[swap] = temp;
                var fitness = calculateKeyFitness(text, key, stat);
                key.fitness = fitness;
            }
        });
    });
    return keys;
}

````python
import itertools

def encontrar_clave(mensaje_cifrado):
    letras_alfabeto = "abcdefghijklmnopqrstuvwxyz"
    mensaje_descifrado = ""
    clave_encontrada = ""
    posibles_claves = list(itertools.permutations(letras_alfabeto)) # Generamos una lista con todas las posibles permutaciones del alfabeto
    
    for clave in posibles_claves:
        diccionario_clave = {}
        mensaje_descifrado = ""

        # Creamos el diccionario que asocia cada letra con su correspondiente en la clave
        for i in range(len(letras_alfabeto)):
            diccionario_clave[letras_alfabeto[i]] = clave[i]

        # Desciframos el mensaje utilizando la clave actual
        for letra in mensaje_cifrado:
            if letra.lower() in letras_alfabeto:
                mensaje_descifrado += diccionario_clave[letra.lower()]
            else:
                mensaje_descifrado += letra

        # Verificamos si el mensaje descifrado es legible
        if mensaje_descifrado.isprintable():
            clave_encontrada = "".join(clave)
            break

    return clave_encontrada
````

````python
from collections import Counter

# Mensaje cifrado
mensaje = "zchdzqtrzdytzirgqstzirzqltgtirsezzdhgqleanmqdlztqreanmqhelnmfqhqldueanmqaqutyuqftehrdhguecdurqlodhnnenmqnliqqzvqlelzdlsirdilqutiradnmqlnedzilgqlqgrehmirbdhgnedzilgqlqgktaqdhgtktuumdoqzcoqhfqdhsqthnmtrutaqelnmqhqyn"

# Análisis de frecuencia de caracteres
frecuencias = Counter(mensaje)

# Diccionario de sustitución
dicc_sustitucion = {}
alfabeto = "abcdefghijklmnopqrstuvwxyz"
for letra in sorted(frecuencias, key=frecuencias.get, reverse=True):
    letra_mas_frecuente = letra
    dicc_sustitucion[letra_mas_frecuente] = alfabeto[0]
    alfabeto = alfabeto[1:]

# Decodificación del mensaje
mensaje_decodificado = ""
for letra in mensaje:
    if letra in dicc_sustitucion:
        mensaje_decodificado += dicc_sustitucion[letra]
    else:
        mensaje_decodificado += letra

print(mensaje_decodificado)
````
