Tipos de objetos 
Beautiful Soup transforma un documento HTML complejo en un árbol complejo de objetos Python. Pero solo tendrás que lidiar con cuatro tipos de objetos: Tag, NavigableString, BeautifulSoupy Comment.

Para analizar un documento, páselo al BeautifulSoup constructor. Puede pasar una cadena o un identificador de archivo abierto:

```python
soup = BeautifulSoup('<b class="boldest">Extremely bold</b>', 'html.parser')
#etiqueta
tag = soup.b
print(type(tag))
print(tag.name)
print(tag)
# <class 'bs4.element.Tag'>
#Primero, el documento se convierte a Unicode y las entidades HTML se convierten a caracteres Unicode:
print(BeautifulSoup("<html><head></head><body>Sacr&eacute; bleu!</body></html>", "html.parser"))
# <html><head></head><body>Sacré bleu!</body></html>
```

Una etiqueta puede tener cualquier número de atributos. La etiqueta tiene un atributo "id" cuyo valor es "más audaz". Puede acceder a los atributos de una etiqueta tratando la etiqueta como un diccionario:<b id="boldest">

Navegando por el árbol 
Bajando 

Las etiquetas pueden contener cadenas y otras etiquetas. Estos elementos son los hijos de la etiqueta . Beautiful Soup proporciona muchos atributos diferentes para navegar e iterar sobre los elementos secundarios de una etiqueta.

Tenga en cuenta que las cadenas Beautiful Soup no admiten ninguno de estos atributos, porque una cadena no puede tener hijos.

Navegando usando nombres de etiquetas ¶
La forma más sencilla de navegar por el árbol de análisis es decir el nombre de la etiqueta que desea. Si desea la etiqueta <head>, simplemente diga soup.head:

```python
from bs4 import BeautifulSoup
with open(variableGlobalPaginaHtml) as pagina:
    soup=BeautifulSoup(pagina, 'html.parser')
#print(soup.head)
print(soup.head.attrs)
print(soup.head.meta)
print(soup.meta.attrs)
print(soup.title)
#print(soup.body)
print(soup.body.img)
print("-----------------------------------------------")
for imagen in soup.find_all('img'):
  print("Imagenes: ",imagen)
print("-----------------------------------------------")
i=0
for referencia in soup.find_all('a'):
  print("Referencia: ",referencia)
  if i>5:
    break
  i+=1
print("-----------------------------------------------")
```

.contentsy .children
Los hijos de una etiqueta están disponibles en una lista llamada .contents:


```python
from bs4 import BeautifulSoup
with open(variableGlobalPaginaHtml) as pagina:
    soup=BeautifulSoup(pagina, 'html.parser')
head_tag = soup.head
print("-----------------------------------------------")
print(head_tag)
print("-----------------------------------------------")
print(head_tag.contents)
print("-----------------------------------------------")
title_tag = head_tag.contents[0]
print(title_tag.contents)
print("-----------------------------------------------")
print(len(title_tag.contents))
#Una cadena no tiene .contents porque no puede contener nada:
```

En lugar de obtenerlos como una lista, puede iterar sobre los hijos de una etiqueta usando el .childrengenerador:

```python
from bs4 import BeautifulSoup
with open(variableGlobalPaginaHtml) as pagina:
    soup=BeautifulSoup(pagina, 'html.parser')
head=soup.title
for hijo in head.children:
  print(hijo)
body=soup.body
i=0
for hijo in body.children:
  print("-----------------------------------------------")
  print("-->: ",hijo)
  print("-----------------------------------------------")
  if i>2:
    break
  i+=1
print(len(body))
```

<!-- .stringsy stripped_strings -->
Si hay más de una cosa dentro de una etiqueta, aún puede mirar solo las cadenas. Usa el .stringsgenerador:

```python
for string in soup.strings:
    print(repr(string))
```

Estas cadenas tienden a tener muchos espacios en blanco adicionales, que puede eliminar utilizando el .stripped_stringsgenerador en su lugar:

```python
for string in soup.stripped_strings:
    print(repr(string))
```

Yendo de lado ¶
La etiqueta <b> y la etiqueta <c> están en el mismo nivel: ambas son elementos secundarios directos de la misma etiqueta. Los llamamos hermanos . Cuando un documento está bastante impreso, los hermanos aparecen en el mismo nivel de sangría. También puede usar esta relación en el código que escribe.

```python
sibling_soup = BeautifulSoup("<a><b>text1</b><c>text2</c></a>", 'html.parser')
#print(sibling_soup.prettify())
print("-----------------------------------------------")
print("---->",sibling_soup.b.next_sibling)
# <c>text2</c>
print("-----------------------------------------------")
print(sibling_soup.c.previous_sibling)
# <b>text1</b>
```

Tipos de filtros ¶
Antes de hablar en detalle sobre find_all()métodos similares, quiero mostrar ejemplos de diferentes filtros que puede pasar a estos métodos. Estos filtros aparecen una y otra vez en toda la API de búsqueda. Puede usarlos para filtrar según el nombre de una etiqueta, sus atributos, el texto de una cadena o alguna combinación de estos.

```python
#filtro por cadenas
from bs4 import BeautifulSoup
with open(variableGlobalPaginaHtml) as pagina:
    soup=BeautifulSoup(pagina, 'html.parser')
print(variableGlobalPaginaHtml)
#print(soup.find_all('br'))
#print(soup.find_all('a'))
#print(soup.find_all('div'))
print(soup.find_all('link'))
listaLink=soup.find_all('link')
print("-----------------------------------------------")
for link in listaLink:
  print("link-->:",link)
  print("atributos-->:",link.attrs)
  print("-----------------------------------------------")
```

```python
#filtro por expresiones regulares para nombre de etiquetas
import re
for tag in soup.find_all(re.compile("^div")):
    print("etiqueta-->:",tag)
    print("nombre-->:",tag.name)
    print("atributos-->:",tag.attrs)
    print("-----------------------------------------------")
```

```python
from bs4 import BeautifulSoup
def contieneAtributos(etiqueta):
  if len(etiqueta.attrs) > 0:
    return True
  else:
    return False
with open(variableGlobalPaginaHtml) as pagina:
    soup=BeautifulSoup(pagina, 'html.parser')
print(variableGlobalPaginaHtml)
for tag in soup.find_all("li"):
  if not contieneAtributos(tag):
    print(tag)
    print("-----------------------------------------------")
```
