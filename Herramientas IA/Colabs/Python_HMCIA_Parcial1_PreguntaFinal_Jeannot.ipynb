```python
from collections import Counter

# Mensaje cifrado
mensaje = "zchdzqtrzdytzirgqstzirzqltgtirsezzdhgqleanmqdlztqreanmqhelnmfqhqldueanmqaqutyuqftehrdhguecdurqlodhnnenmqnliqqzvqlelzdlsirdilqutiradnmqlnedzilgqlqgrehmirbdhgnedzilgqlqgktaqdhgtktuumdoqzcoqhfqdhsqthnmtrutaqelnmqhqyn"

# Análisis de frecuencia de caracteres
frecuencias = Counter(mensaje)

# Diccionario de sustitución
dicc_sustitucion = {}
alfabeto = "abcdefghijklmnopqrstuvwxyz"
for letra in sorted(frecuencias, key=frecuencias.get, reverse=True):
    letra_mas_frecuente = letra
    dicc_sustitucion[letra_mas_frecuente] = alfabeto[0]
    alfabeto = alfabeto[1:]

# Decodificación del mensaje
mensaje_decodificado = ""
for letra in mensaje:
    if letra in dicc_sustitucion:
        mensaje_decodificado += dicc_sustitucion[letra]
    else:
        mensaje_decodificado += letra

print(mensaje_decodificado)
```
