## **seaborn.relplot**
Esta función brinda acceso a varias funciones a nivel de ejes diferentes que muestran la relación entre dos variables con asignaciones semánticas de subconjuntos. El parámetro **kind** selecciona la función de nivel de ejes subyacente para usar:

- scatterplot() (con kind="scatter"; por defecto)

- lineplot() (con kind="line")



```python
seaborn.relplot(data=None, x=None, y=None, col=None, kind='scatter')
```

````python
#Gráfica de datos
sns.relplot(data=iris, x="sepal_length", y="petal_length",col="species")
plt.show()
````

````python
sns.relplot(data=iris, x="sepal_length", y="petal_length",hue="species", 
            style="species");
````

La función *relplot()*, facilita la representación de datos por medio de un **diagrama de dispersion**, sin embargo esto se puede mejorar con una linea para esto existe el parametro **kind**.

````python
sns.relplot(data=iris, x="sepal_length", y="petal_length",col="species", 
            kind="line")
plt.show()
````

# Estimación estadística 

### *seaborn.lmplot* 
La estimación estadística en seaborn no es solo estadistica descriptiva. Por ejemplo, es posible mejorar un diagrama de dispersión al incluir un modelo de regresión lineal (y su incertidumbre) usando lmplot().



```python
seaborn.lmplot(data=None, x=None, y=None, col=None)
```





````python
sns.lmplot(data=iris, x="sepal_length", y="petal_length",col="species")
plt.show()
````

## **seaborn.displot**
Si lo que se desea es visualizar una representación distributiva se puede usar la función displot().
Esta función brinda acceso a varios enfoques para visualizar la distribución de datos univariada o bivariada, incluidos subconjuntos de datos definidos por mapeo semántico y facetas en varias subparcelas. El parámetro **kind** selecciona el enfoque a utilizar:

- histplot() (con kind="hist"; por defecto)

- kdeplot() (con kind="kde")

- ecdfplot() (con kind="ecdf"; solo univariado)



```python
seaborn.displot(data=None, x=None, y=None, col=None, hue=None, kind='hist', rug=None)
```



````python
#Traza distribuciones univariadas o bivariadas utilizando la estimación de 
#densidad kernel.
sns.displot(data=iris, x="sepal_length", col="species", kde=True)
plt.show()
````

````python
#Traza funciones de distribución acumulativa empírica.
#rug --> Traza una marca en cada valor de observación a lo largo de los ejes x 
#y/o y.
sns.displot(data=iris, x="sepal_length",col="species", kind="ecdf", rug=True)
plt.show()
````

````python
#Gráfica todas las distribuciones de las especies en un solo plot y diferencia
# por color
sns.displot(data=iris, x="sepal_length", hue="species", kde=True)
plt.show()
````

````python
#Distribución de las especies al correlacionar dos atributos
sns.displot(data=iris, x="sepal_length", y="petal_length", hue="species")
plt.show()
````

````python
#Gráfica histograma, es posible modificar el numero de barras (bins) y el modo 
#de la distribución (acumulativo)
sns.histplot(data=iris, x="sepal_width", bins=5, cumulative=True)
plt.show()
````

````python
#Cambia la variable y del histograma
sns.histplot(data=iris, x="petal_width", bins=5, stat="percent", 
             multiple="dodge", hue="species")
plt.show()
````

## **seaborn.catplot**
Esta función brinda acceso a varias funciones a nivel de ejes que muestran la relación entre una variable numérica y una o más categóricas usando una de varias representaciones visuales. El parámetro **kind** selecciona la función de nivel de ejes subyacente para usar:

Gráficos para datos categóricos:

- Diagramas de dispersión categóricos:

  - stripplot() (con kind="strip"; por defecto)

  - swarmplot() (con kind="swarm")

- Diagramas de distribución categórica:

  - boxplot() (con kind="box")

  - violinplot() (con kind="violin")

  - boxenplot() (con kind="boxen")

- Gráficos de estimación categórica:

  - pointplot() (con kind="point")

  - barplot() (con kind="bar")

  - countplot() (con kind="count") 



```python
seaborn.catplot(data=None, x=None, y=None, col=None, kind='strip', split=None)
```



````python
sns.catplot(data=iris, x="sepal_length", y="species", kind="swarm")
plt.show()
````

````python
# con el argumento kind="violin" y split=True, se puede representar la 
#distribución subyacente de la que se muestrean los puntos
sns.catplot(data=iris, x="sepal_length", y="species", kind="violin", split=True)
plt.show()
````

````python
# alternativamente con el argumento kind='bar', podemos crear un grafico de
# barras que en este caso nos muestra su valor medio e intervalo de confianza
sns.catplot(data=iris, x="sepal_length", y="species", kind="bar")
plt.show()
````

````python
#Grafica diagrama de cajas y lo combina con un swarmplot
plt.figure(figsize=(10,10))
sns.boxplot(data=iris, y="sepal_length", x="species");
sns.swarmplot(data=iris, y="sepal_length", x="species", color="0", marker="<");
````

## **seaborn.jointplot**

Dibuja una gráfica de dos variables con gráficos bivariados y univariados.



```python
seaborn.jointplot(data=None, x=None, y=None, hue=None, kind='scatter', marginal_ticks=False)
```



````python
#Gráfica dos diagramas diferente en el mismo plot
sns.jointplot(data=iris, x="petal_length", y="petal_width", hue="species", 
              marginal_ticks=True);
````

## **seaborn.pairplot**
Traza relaciones por pares en un conjunto de datos.




````python
#Gráfica las variable del dataset emparedas
sns.pairplot(data=iris, hue="species");
````
